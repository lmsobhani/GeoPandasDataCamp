```python
import geopandas as gpd
```

```python
# Load the GeoJSON file from the URL into a GeoDataFrame.
url = "https://raw.githubusercontent.com/jcanalesluna/bcn-geodata/master/districtes/districtes.geojson"
districts = gpd.read_file(url)
#districts # Displays the GeoDataFrame
```

```python
# Save the GeoDataFrame to a GeoJSON file.
districts.to_file("districts.geojson", driver="GeoJSON")
```

```python
districts.crs # Displays the Coordinate Reference System (CRS) of the GeoDataFrame
```

```python
districts.to_crs(epsg=2062, inplace=True) # Reproject the GeoDataFrame to EPSG:2062
districts.crs # Displays the new Coordinate Reference System (CRS) of the GeoDataFrame
```

```python
#Attribute 1 - Area
districts["area"] = districts.area/1000000  # Calculate area in square kilometers
#districts
```

```python
#Attribute 2 - Centroid
districts["centroid"]=districts.centroid
#districts
```

```python
#Attribute 3 - Boundary
districts["boundary"]=districts.boundary
#districts
```

```python
#Attribute 4 - Distance
#gives min distance from geometry to location.
#Step 1: Create a point using the shapely Point() function w/ desired coordinates. 
# Then, convert it into a GeoSeries w/ the CRS.

from shapely.geometry import Point

sagrada_fam = Point( 2.174356, 41.403629 ) # Coordinates of La Sagrada Familia Church
sagrada_fam = gpd.GeoSeries(sagrada_fam, crs=4326)
sagrada_fam = sagrada_fam.to_crs(epsg=2062)
```
